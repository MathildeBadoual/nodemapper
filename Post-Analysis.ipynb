```python
def mape(errors):
    return np.mean(np.absolute(errors))

mape([1,2,3])
```

```python
new_linear_models = pd.read_pickle('new_linear_results.p')
new_nonlinear_models = pd.read_pickle('new_nonlinear_results.p')
```

```python
all_linear_models = pd.concat([linear_models, new_linear_models]).reset_index(drop=True)
all_nonlinear_models = pd.concat([nonlinear_models, new_nonlinear_models]).reset_index(drop=True)
```

```python
numbers = [10, 20, 3]
names = ['Regression (All Data)', 'Regression (Lags Only)', 'Elastic Net (All Data)']
for_graph = all_linear_models.iloc[numbers]
for_graph['Model'] = names

sns.barplot(x="score", y="Model", data=for_graph)
```

```python
nonlinear_models['Model'] = ['Random Forest (No Lags)',
                            'Gradient Boost (No Lags)',
                            'Random Forest (All Variables)',
                            'Gradient Boost (All Variables)']

all_nonlinear = pd.concat([nonlinear_models, new_nonlinear_models])

all_nonlinear['Model'] = ['Random Forest (No Lags)',
                          'Gradient Boost (No Lags)',
                          'Random Forest (All Variables)',
                          'Gradient Boost (All Variables)',
                          'Random Forest (Lags Only)',
                          'Gradient Boost (Lags Only)']

sns.barplot(x="score", y="Model", data=all_nonlinear)
```

```python
all_graph_data = pd.concat([all_nonlinear, for_graph])
all_graph_data.to_csv("graph_data.csv")
```

```python
def compare_errors(model1, model2, df):
    errors1 = df[df['Model']==model1]['errors'].values[0][0]
    errors2 = df[df['Model']==model2]['errors'].values[0][0]
    df = pd.DataFrame()
    df[model1] = errors1
    df[model2] = errors2
    print(ggplot(aes(x = model1, y = model2), data = df) + geom_point())
    
data = compare_errors('Regression (Lags Only)', 'Elastic Net (All Data)', all_graph_data)
```
